```python
import numpy as np
from time import time
import MOD_CG
from scipy.io import mmread
from scipy.sparse import coo_matrix
from numpy import linalg as lg
from tabulate import tabulate
```

```python
A2=np.array(mmread('content/bcsstm19.mtx.gz').todense())
A3=np.array(mmread('content/bcsstm20.mtx.gz').todense())
C3=np.array(mmread('content/bcsstm26.mtx.gz').todense())
B1=np.array(mmread('content/bcsstm21.mtx.gz').todense())
C1=np.array(mmread('content/bcsstm22.mtx(1).gz').todense())
B2=np.array(mmread('content/bcsstk16.mtx.gz').todense())


```

```python
matrices=[A2,A3,B1,B2,C1,C3]
tol=1e-5
itermax = 50000
for i in range(len(matrices)):
    if (np.all(np.linalg.eigvals(matrices[i]) < 0)):
       print('Oups, something went wrong')
       break
print('if only this message appears, comgrats! all the matrices are positive definite')

```

```python
my_data = []
for i in range(len(matrices)):
  l = []
  n = len(matrices[i])
  b=np.ones(n)
  x = np.zeros(n)
  t1=time()
  iter = MOD_CG.CG(matrices[i],b,x,itermax,tol) 
  t2=time()
  tmp=(t2-t1)*1000
  res=lg.norm(np.dot(matrices[i],x)-b)  
  l.append(matrices[i].shape)
  l.append(lg.cond(matrices[i]))
  l.append(tol)
  l.append(res)
  l.append(iter)
  l.append(tmp)
  
  my_data.append(l)
# create header
head = ['Shape','conditio number','tol','res','iter','temp(ms)']
  
# display table
print(tabulate(my_data, headers=head, tablefmt="grid")) 


```
