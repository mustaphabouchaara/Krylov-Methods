```python
import numpy as np
from time import time
# MOD_BICG refers to the .so file after executing pyccel on MOD_BICG.py
import MOD_BICG
from scipy.io import mmread
from scipy.sparse import coo_matrix
from numpy import linalg as lg
from tabulate import tabulate

```

```python
###################################

A2 = np.array(mmread('content/impcol_c.mtx.gz').todense())
C5 = np.array(mmread('content/sherman4.mtx.gz').todense())
C6 = np.array(mmread('content/orsreg_1.mtx.gz').todense())
C7 = np.array(mmread('content/orsirr_1.mtx.gz').todense())
C1 = np.array(mmread('content/west0067.mtx.gz').todense())
C9 = np.array(mmread('content/gre__185.mtx.gz').todense())
C13 = np.array(mmread('content/e05r0200.mtx.gz').todense())
B4 = np.array(mmread('content/bcsstm24.mtx.gz').todense())

###################################

```

```python
matrices = [A2, C5, C6, C7, C1, C9, C13]
tol = 1e-5
itermax = 50000
```

```python
my_data = []
for i in range(len(matrices)):
    l = []
    n = len(matrices[i])
    b = np.ones(n)
    x = np.zeros(n)
    t1 = time()
    iter = MOD_BICG.BICG(matrices[i], b, x, tol, itermax)
    t2 = time()
    tmp = (t2-t1)*1000
    res = lg.norm(np.dot(matrices[i], x)-b)
    l.append(matrices[i].shape)
    l.append(lg.cond(matrices[i]))
    l.append(tol)
    l.append(res)
    l.append(iter)
    l.append(tmp)

    my_data.append(l)
# create header
head = ['Shape', 'conditio number', 'tol', 'res', 'iter', 'temp(ms)']

# display table
print(tabulate(my_data, headers=head, tablefmt="grid"))
```
